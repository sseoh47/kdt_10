```python
## ===============================================================
## [문제 2] 반별 평균을 통합할 때 – 가중평균
## ===============================================================
## A반과 B반의 시험 결과가 다음과 같다.
##
## A반: 학생 수 10명, 평균 90점
##
## B반: 학생 수 40명, 평균 70점

## 전체 50명에 대한 단순 평균을 잘못 계산한 경우 (반 평균의 평균)과
##
simple_avg=90+70/50
print(f'단순평균 => {simple_avg}')

## 올바른 전체 평균(가중평균) 을 계산하세요.
weighted_avg=(90*10+40*70)/50
print(f"가중평균 => {weighted_avg}")
```

```python
## ===============================================================
## [문제 3] 절사 평균 계산
## ===============================================================
## 데이터는 아래와 같습니다.
x=[10,12,11,9,13,200]

xSR=pd.Series(x)
x
## 일반 평균을 구하시오.
print(f"일반 평균 => {xSR.mean()}")
## 가장 작은 값과 가장 큰 값(10, 200)을 하나씩 제거한 후 절사 평균을 구하시오.
print(f"절사 평균 => {xSR[1:-1].mean()}")
## 왜 절사 평균이 일반 평균보다 극단값에 덜 민감한지 설명하시오.
# : 일반평균은 모든 값을 동일한 비중으로 계산 => 극단값 있을 경우 평균 왜곡된다.
#   그러나 절사 평균은 (일정 수의)극단값을 제거한 후 계산하는 방식이므로 극단값에 덜 민감하다.
```

```python
## ===============================================================
## [문제 4] 가중 평균 계산
## ===============================================================
## 어떤 회사의 고객 만족도 조사는 다음과 같습니다.
##  그룹	응답자 수	평균 점수
##  VIP	       20	    4.8
##  일반	  200	    4.2
##
df=pd.DataFrame({
    "그룹":["VIP", "일반"],
    "응답자 수":[20,200],
    "평균 점수":[4.8, 4.2]
})
display(df)

## 전체 고객의 가중 평균 만족도 점수를 계산하시오.
weighted_avg=np.average(df["평균 점수"],weights=df["응답자 수"])
print(f"가중평균 만족도 점수 => {weighted_avg}")
## 단순히 (4.8 + 4.2) / 2 로 계산하는 것이 왜 잘못되었는지 설명하시오.
# : 두 그룹의 응답자 수는 다르기에 단순히 계산했을 때는 1:1 계산이 된다.
#   -> VIP가 적은만큼 점수가 과대 반영되는 결과가 나타나게 되며 결과에 오류가 발생한다.
```

```python
## ===============================================================
##  [문제 5] Pandas로 그룹별 가중평균 계산
## ===============================================================
## 다음과 같은 DataFrame이 있습니다.

df = pd.DataFrame({
    "group":  ["A", "A", "A", "B", "B"],
    "score":  [80, 90, 70, 60, 100],
    "weight": [1, 2, 1, 3, 1]
})
display(df)

## group별 단순 평균(score의 평균)을 구하시오.
simple_avg=df.groupby("group")["score"].mean()
print(f"단순평균 => {simple_avg}")

## group별 가중평균(weight를 가중치로 사용)을 구하는 Pandas 코드를 작성하시오.
df.groupby("group").apply(lambda x:np.average(df["score"], weights=df["weight"]))
```

```python
## ===============================================================
##  [ 문제 6 ] 분산과 표준편차 비교
## ===============================================================
## 다음은 어느 반 학생 10명의 수학 시험 점수이다.
A = [70, 72, 75, 75, 76, 80, 82, 85, 88, 90]
B = [40, 55, 60, 70, 75, 90, 95, 98, 100, 100]
aSR=pd.Series(A)
bSR = pd.Series(B)

## 두 반의 분산(sample variance)을 각각 구하시오.
A_var = aSR.var()
B_var = bSR.var()
print(f"A반 표본분산 => {A_var:.2f}")
print(f"B반 표본분산 => {B_var:.2f}")
## Pandas Series.var() : 기본값: ddof=1/표본분산
## NumPy np.var() : 기본값: ddof=0/모분산

## 두 반의 표준편차를 비교하고, 어느 반의 점수 분포가 더 불안정한지 설명하시오.
A_std = aSR.std()
B_std = bSR.std()
print(f"A반 표준편차 => {A_std:.2f}")
print(f"B반 표준편차 => {B_std:.2f}")
```
